```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split # to split dataset into train and test
import sys
sys.path.append("..")
from scripts.prepare_data import standarization_unit_variance, normalize
```

```python
##### Reading Data #####

DATASET_PATH = "../datasets/dataset.csv"
# Get dataset
df = pd.read_csv(DATASET_PATH, sep=",")

# cols = df.columns[df.columns.isin(['emotion'])]
# df = df[(df[cols] == 0).all(1)] # Only desired emotion

# normal [3:16]
X = df[df.columns[3,6]] # Only the MFCC features
y = df[df.columns[-1]] # Emotion label

# Normalization of input features in X
# X = normalize(X)
# X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.1)

```

```python
len(y) == len(X)
```

```python
kmeans = KMeans(n_clusters=7, random_state=0).fit(X)
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit(X)
pca_2d = pca.transform(X)
```

```python
pca_2d = X.values
```

```python
colors = [
    '#555E7B',
    '#B7D968',
    '#B576AD',
    '#E04644',
    '#FDE47F',
    '#7CCCE5',
    '#130FFC'
]
```

```python
import pylab as pl
for i in range(0, pca_2d.shape[0]):
    if y[i] == 0:
        c1 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c=colors[0], marker='o')
    elif y[i] == 1:
        c2 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c=colors[1], marker='o')
    elif y[i] == 2:
        c3 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c=colors[2], marker='o')
    elif y[i] == 3:
        c4 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c=colors[3], marker='o')
    elif y[i] == 4:
        c5 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c=colors[4], marker='o')
    elif y[i] == 5:
        c6 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c=colors[5], marker='o')
    elif y[i] == 6:
        c7 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c=colors[6], marker='o')
 
pl.legend([c1, c2, c3, c4, c5, c6, c7], ['Alegria', 'Nojo', 'Medo', 'Neutro', 'Raiva', 'Surpresa', 'Tristeza'])
pl.title('Features X Classification')
pl.show()
```
